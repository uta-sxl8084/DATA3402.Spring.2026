*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
def draw_board(board):
    n = len(board)

    for i in range(n):
        print(" ---" * n + " ")
        row = ""
        for j in range(n):
            cell = board[i][j]

            if cell == input_x:
                symbol = "X"
            elif cell == input_o:
                symbol = "O"
            else:
                symbol = " "

            row += "| " + symbol + " "
        print(row + "|")

    print(" ---" * n + " ")
````

````python
board = create_board(3)

board[0][0] = input_x
board[1][1] = input_o
board[2][2] = input_x

draw_board(board)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
def check_winner(board):
    n = len(board)
    has_empty = False

    for row in board:
        if row.count(row[0]) == n and row[0] != 0:
            return row[0]
        if 0 in row:
            has_empty = True

    for col in range(n):
        column = [board[row][col] for row in range(n)]
        if column.count(column[0]) == n and column[0] != 0:
            return column[0]

    diag1 = [board[i][i] for i in range(n)]
    if diag1.count(diag1[0]) == n and diag1[0] != 0:
        return diag1[0]

    diag2 = [board[i][n - i - 1] for i in range(n)]
    if diag2.count(diag2[0]) == n and diag2[0] != 0:
        return diag2[0]

    if has_empty:
        return -1  
    else:
        return 0   
````

````python
winner_is_2 = [[2, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_1 = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 1]]

winner_is_also_1 = [[0, 1, 0],
	[2, 1, 0],
	[2, 1, 1]]

no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 2]]

also_no_winner = [[1, 2, 0],
	[2, 1, 0],
	[2, 1, 0]]
````

````python
print(check_winner(winner_is_2))        
print(check_winner(winner_is_1))        
print(check_winner(winner_is_also_1))   
print(check_winner(no_winner))          
print(check_winner(also_no_winner))     
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
def make_move(board, player, row, col):
    n = len(board)

    if row < 0 or row >= n or col < 0 or col >= n:
        return False

    if board[row][col] != 0:
        return False

    board[row][col] = player
    return True
````

````python
board = create_board(3)

print(make_move(board, 1, 0, 0))  
print(make_move(board, 2, 0, 0))  
print(make_move(board, 2, 3, 3)) 

draw_board(board)
````

*Exercise 6:* Modify Exercise 3 to show **row and column labels** so that players can specify locations like `"A2"` or `"C1"`.

````python
def draw_board_labeled(board):
    n = len(board)

    print("    " + "   ".join(str(i+1) for i in range(n)))

    for r in range(n):
        print("    " + "--- " * n)

        row_label = chr(ord('A') + r)
        print(f"{row_label}  ", end="")

        for c in range(n):
            cell = board[r][c]
            if cell == 1:
                symbol = "X"
            elif cell == 2:
                symbol = "O"
            else:
                symbol = " "
            print(f"| {symbol} ", end="")
        print("|")

    print("    " + "--- " * n)
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
def make_move_from_input(board, player, move):
    move = move.strip().upper()

    n = len(board)

    if len(move) < 2:
        return False

    row_char = move[0]
    col_part = move[1:]

    row = ord(row_char) - ord('A')

    if not col_part.isdigit():
        return False
    col = int(col_part) - 1

    return make_move(board, player, row, col)
````

````python
board = create_board(3)

print(make_move_from_input(board, 1, "A1"))
print(make_move_from_input(board, 2, "C2"))

draw_board_labeled(board)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
def player_turn(board, player):
    while True:
        move = input(f"Player {player} enter your move (e.g., A1): ")

        if make_move_from_input(board, player, move):
            break
        else:
            print("Invalid move. Try again.")
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
def play_tic_tac_toe(size):
    board = create_board(size)
    current_player = 1

    while True:
        draw_board_labeled(board)
        print(f"Player {current_player}'s turn")

        player_turn(board, current_player)

        result = check_winner(board)

        if result == 1:
            draw_board_labeled(board)
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            draw_board_labeled(board)
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            draw_board_labeled(board)
            print("It's a draw!")
            break

        current_player = 2 if current_player == 1 else 1
````

````python
play_tic_tac_toe(3)
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
def find_winning_move(board, player):
    n = len(board)

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                board[r][c] = player
                if check_winner(board) == player:
                    board[r][c] = 0
                    return (r, c)
                board[r][c] = 0
    return None

def computer_turn(board, computer_player):
    human_player = 1 if computer_player == 2 else 2

    move = find_winning_move(board, computer_player)
    if move:
        make_move(board, computer_player, move[0], move[1])
        return

    move = find_winning_move(board, human_player)
    if move:
        make_move(board, computer_player, move[0], move[1])
        return

    n = len(board)
    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                make_move(board, computer_player, r, c)
                return

def play_vs_computer():
    board = create_board(3)
    human = 1
    computer = 2
    current_player = 1

    while True:
        draw_board_labeled(board)

        if current_player == human:
            print("Your turn!")
            player_turn(board, human)
        else:
            print("Computer's turn :|")
            computer_turn(board, computer)

        result = check_winner(board)

        if result == human:
            draw_board_labeled(board)
            print("You win!")
            break
        elif result == computer:
            draw_board_labeled(board)
            print("Computer wins!")
            break
        elif result == 0:
            draw_board_labeled(board)
            print("It's a draw!")
            break

        current_player = 2 if current_player == 1 else 1
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
def computer_turn_search(board, computer_player, max_depth):
    human_player = 1 if computer_player == 2 else 2

    def minimax(board, depth, is_maximizing):
        result = check_winner(board)

        if result == computer_player:
            return 10 - depth
        elif result == human_player:
            return depth - 10
        elif result == 0:
            return 0

        if depth == max_depth:
            return 0

        n = len(board)

        if is_maximizing:
            best_score = -float('inf')
            for r in range(n):
                for c in range(n):
                    if board[r][c] == 0:
                        board[r][c] = computer_player
                        score = minimax(board, depth + 1, False)
                        board[r][c] = 0
                        best_score = max(best_score, score)
            return best_score
        else:
            best_score = float('inf')
            for r in range(n):
                for c in range(n):
                    if board[r][c] == 0:
                        board[r][c] = human_player
                        score = minimax(board, depth + 1, True)
                        board[r][c] = 0
                        best_score = min(best_score, score)
            return best_score

    best_score = -float('inf')
    best_move = None
    n = len(board)

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                board[r][c] = computer_player
                score = minimax(board, 1, False)
                board[r][c] = 0

                if score > best_score:
                    best_score = score
                    best_move = (r, c)

    if best_move:
        make_move(board, computer_player, best_move[0], best_move[1])
````

````python
board = [
    [2, 2, 0],
    [1, 1, 0],
    [0, 0, 0]
]

draw_board_labeled(board)
computer_turn_search(board, 2, max_depth=4)
print("After computer move:")
draw_board_labeled(board)
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
def play_cvc(size, depth_smart, depth_dumb):
    board = create_board(size)
    current_player = 1

    while True:
        if current_player == 1:
            computer_turn_search(board, 1, depth_smart)
        else:
            computer_turn_search(board, 2, depth_dumb)

        result = check_winner(board)
        if result != -1:
            return result

        current_player = 2 if current_player == 1 else 1

def run_tournament(size, games, depth_smart, depth_dumb):
    smart_wins = 0
    dumb_wins = 0
    draws = 0

    for i in range(games):
        result = play_cvc(size, depth_smart, depth_dumb)
        if result == 1:
            smart_wins += 1
        elif result == 2:
            dumb_wins += 1
        else:
            draws += 1

    print(f"\nBoard Size: {size}x{size}")
    print(f"Smart AI wins: {smart_wins}")
    print(f"Dumb AI wins: {dumb_wins}")
    print(f"Draws: {draws}")
    print(f"Smart win rate: {smart_wins/games*100:.1f}%")
````

````python
run_tournament(3, games=10, depth_smart=6, depth_dumb=1)
run_tournament(4, games=10, depth_smart=3, depth_dumb=1)
run_tournament(5, games=10, depth_smart=2, depth_dumb=1)
````
